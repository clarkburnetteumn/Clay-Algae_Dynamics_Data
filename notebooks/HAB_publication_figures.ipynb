```python
# Tables 1 and 2
def mean_histogram_shape(
    filepath,
    sample_indices,
    condition="Opt",
    depth_range=None,
    normalize=False,
    plot=False, 
    title=''
):
    """
    Compute mean histogram shape for a given dosage condition.

    Parameters
    ----------
    filepath : str
        CSV file path

    sample_indices : list
        Sampling window indices

    condition : str
        One of ["Baseline", "Low", "Opt", "High", "Mix"]

    depth_range : tuple, optional
        Restrict averaging to (min_depth, max_depth)

    normalize : bool, optional
        If True, normalize each histogram before averaging

    plot : bool
        Whether to plot the result

    Returns
    -------
    mean_hist : ndarray
        Mean histogram (length = number of bins)
    """

    df = pd.read_csv(filepath, header=None)

    bin_cols = list(range(36))
    depth_col = 40
    volume_col = 50

    group_labels = ["Baseline", "Low", "Opt", "High", "Mix"]
    tanks_per_condition = 3
    profiles_per_tank = 2

    # --- Build condition runs ---
    sampling_runs = list(zip(sample_indices[0::2], sample_indices[1::2]))

    condition_runs = {}
    for i, label in enumerate(group_labels):
        start = i * tanks_per_condition * profiles_per_tank
        end = start + tanks_per_condition * profiles_per_tank
        condition_runs[label] = sampling_runs[start:end]

    runs = condition_runs[condition]

    # --- Collect histograms ---
    histograms = []
    tvc_values = []

    for run_start, run_end in runs:

        profile = df.iloc[run_start:run_end + 1]
        depth0 = profile[depth_col].min()

        for _, row in profile.iterrows():

            depth = row[depth_col] - depth0

            if depth_range is not None:
                if not (depth_range[0] <= depth <= depth_range[1]):
                    continue

            H = row[bin_cols].values.astype(float)
            H = np.nan_to_num(H, nan=0.0)

            V = pd.to_numeric(row[volume_col], errors='coerce')

            if normalize:
                total = np.sum(H)
                if total > 0:
                    H = H / total

            if np.isfinite(V):
                tvc_values.append(V)
            histograms.append(H)

    if len(histograms) == 0:
        raise ValueError("No valid histograms found for given condition/depth range")

    histograms = np.array(histograms)

    # --- Mean shape ---
    mean_hist = np.nanmean(histograms, axis=0)

    # --- 95% confidence interval ---
    n_hist = len(histograms)
    std_hist = np.nanstd(histograms, axis=0, ddof=1)
    sem_hist = std_hist / np.sqrt(n_hist)
    ci95_hist = 1.96 * sem_hist



    # --- TVC ---
    tvc_values = np.array(tvc_values)

    mean_tvc = np.nanmean(tvc_values) if len(tvc_values) > 0 else np.nan
    std_tvc = np.nanstd(tvc_values) if len(tvc_values) > 0 else np.nan

    # --- Plot ---
    if plot:
        import matplotlib.pyplot as plt

        bin_centers = np.logspace(np.log10(1), np.log10(500), len(mean_hist))

        plt.figure(figsize=(12,4))
        plt.bar(bin_centers, mean_hist, width=np.diff(bin_centers, prepend=bin_centers[0]), color='white', edgecolor='black')
        plt.xscale('log')

        # ✅ Force correct tick labels
        tick_values = [1, 5, 25, 100, 500]
        plt.xticks(tick_values, labels=[str(v) for v in tick_values])

        plt.xlabel("Particle Size (μm)")
        plt.ylabel("Mean Volume Concentration (μL/L)")

        if normalize:
            title += " (Normalized)"
        plt.title(title)

        plt.grid(True, which='major', alpha = 0.3, linewidth=0.5)
        plt.tight_layout()
        plt.show()

    return mean_hist, ci95_hist, mean_tvc, std_tvc

def specific_depth_averaged_histograms(
    filepath_t1,
    filepath_t2,
    sample_indices_t1,
    sample_indices_t2,
    depth_range=None,
    clip_percentile=None,
    condition=None,
    vmax=None
):
    """
    Parameters
    ----------
    filepath_t1, filepath_t2 : str
        CSV file paths

    sample_indices_t1, sample_indices_t2 : list
        Sampling windows

    depth_range : tuple, optional
        Restrict depth region (min, max)

    clip_percentile : float, optional
        Remove outliers (e.g., 95 keeps middle 95%)

    condition : str or list, optional
        Select subset of conditions

    vmax : float, optional
        Fix x-axis max
    """

    df1 = pd.read_csv(filepath_t1, header=None)
    df2 = pd.read_csv(filepath_t2, header=None)

    bin_cols = list(range(36))
    depth_col = 40
    volume_col = 50

    bin_widths = np.array([
        0.48, 0.26, 0.31, 0.37, 0.44, 0.52, 0.60, 0.72, 0.85, 1,
        1.17, 1.4, 1.68, 1.9, 2.3, 2.7, 3.2, 3.7, 4.5, 5.2,
        6.2, 7.2, 8.6, 10.2, 11.9, 14.2, 16.4, 20, 23, 28,
        32, 38, 47, 57, 66, 80
    ])

    group_labels = ["Baseline", "Low", "Opt", "High", "Mix"]

    # --- Condition filter ---
    if condition is not None:
        if isinstance(condition, str):
            condition = [condition]
        group_labels = [c for c in group_labels if c in condition]

    tanks_per_condition = 3
    profiles_per_tank = 2

    def build_condition_runs(sample_indices):
        runs = list(zip(sample_indices[0::2], sample_indices[1::2]))
        cond_dict = {}
        for i, label in enumerate(["Baseline", "Low", "Opt", "High", "Mix"]):
            start = i * tanks_per_condition * profiles_per_tank
            end = start + tanks_per_condition * profiles_per_tank
            cond_dict[label] = runs[start:end]
        return cond_dict

    cond_runs_t1 = build_condition_runs(sample_indices_t1)
    cond_runs_t2 = build_condition_runs(sample_indices_t2)

    def compute_profiles(df, cond_runs):

        histograms = []
        tvc_values = []

        for label in group_labels:

            runs = cond_runs[label]

            for run_start, run_end in runs:

                profile = df.iloc[run_start:run_end + 1]
                depth0 = profile[depth_col].min()

                for _, row in profile.iterrows():

                    hist = row[bin_cols].values.astype(float)
                    hist = np.nan_to_num(hist, nan=0.0)
                    V = pd.to_numeric(row[volume_col], errors='coerce')

                    depth = row[depth_col] - depth0

                    if depth_range is not None:
                        if not (depth_range[0] <= depth <= depth_range[1]):
                            continue
                    if np.isfinite(V):
                        tvc_values.append(V)
                    histograms.append(hist)

        histograms = np.asarray(histograms)

        if clip_percentile is not None and len(histograms) > 0:

            volumes = np.sum(histograms * bin_widths, axis=1)

            lo, hi = np.percentile(
                volumes,
                [(100 - clip_percentile)/2,
                100 - (100 - clip_percentile)/2]
            )

            mask = (volumes >= lo) & (volumes <= hi)
            histograms = histograms[mask]

        if len(histograms) == 0:
            return np.full(36, np.nan)
        # --- TVC ---
        tvc_values = np.array(tvc_values)

        mean_tvc = np.nanmean(tvc_values) if len(tvc_values) > 0 else np.nan

        return np.nanmean(histograms, axis=0), mean_tvc

    hist_t1, tvc_t1 = compute_profiles(df1, cond_runs_t1)
    hist_t2, tvc_t2 = compute_profiles(df2, cond_runs_t2)
    

    return hist_t1, tvc_t1, hist_t2, tvc_t2

def summarize_histogram(
    hist, tvc,
    normalize=True,
    size_ranges=None
):
    """
    Compute summary statistics from a histogram.

    Parameters
    ----------
    hist : array-like
        Bin heights

    bin_centers : array-like
        Representative value for each bin (same length as hist)

    normalize : bool
        If True, normalize histogram to sum to 1

    size_ranges : list of tuples, optional
        e.g. [(0,5), (5,25), (25,100), (100, np.inf)]

    Returns
    -------
    stats : dict
        Dictionary of summary statistics
    """

    import numpy as np

    hist = np.array(hist, dtype=float)
    bin_centers = [1.21, 1.60, 1.89, 2.23, 2.63, 3.11,
                   3.67, 4.33, 5.11, 6.03, 7.11, 8.39,
                   9.90, 11.7, 13.8, 16.3, 19.2, 22.7,
                   26.7, 31.6, 37.2, 43.9, 51.9, 61.2,
                   72.2, 85.2, 101, 119, 140, 165,
                   195, 230, 273, 324, 386, 459]
    bin_centers = np.array(bin_centers, dtype=float)

    # --- Clean ---
    hist = np.nan_to_num(hist, nan=0.0)

    total = np.sum(hist)
    if total == 0:
        raise ValueError("Histogram is empty")

    if normalize:
        hist = hist / total

    # --- CDF ---
    cdf = np.cumsum(hist)

    # --- Percentiles ---
    def percentile(p):
        return np.interp(p, cdf, bin_centers)

    D10 = percentile(0.10)
    D50 = percentile(0.50)
    D90 = percentile(0.90)

    # --- Mean ---
    mean = np.sum(hist * bin_centers)

    # --- Std dev ---
    variance = np.sum(hist * (bin_centers - mean)**2)
    std = np.sqrt(variance)

    # --- Skewness ---
    skew = np.sum(hist * ((bin_centers - mean)/std)**3) if std > 0 else np.nan

    # --- Span ---
    span = (D90 - D10) / D50 if D50 != 0 else np.nan

    stats = {
        "D10": D10,
        "D50": D50,
        "D90": D90,
        "mean": mean,
        "std": std,
        "span": span,
        "skewness": skew,
        "TVC": tvc
    }

    # --- Size fractions ---
    if size_ranges is not None:
        for (lo, hi) in size_ranges:
            mask = (bin_centers >= lo) & (bin_centers < hi)
            frac = np.sum(hist[mask])
            stats[f"frac_{lo}_{hi}"] = frac

    return stats

def mean_histogram_shape_simple(
    filepath,
    row_selection,
    depth_range=None,
    normalize=False,
    plot=False, title=''
):
    """
    Compute mean histogram shape from selected rows.

    Parameters
    ----------
    filepath : str
        CSV file path

    row_selection : list or tuple
        Either:
        - [start, end] for a continuous range
        - list of row indices

    depth_range : tuple, optional
        Restrict averaging to (min_depth, max_depth)

    normalize : bool
        If True, normalize each histogram before averaging

    plot : bool
        Whether to plot the result

    Returns
    -------
    mean_hist : ndarray
    mean_tvc : float
    std_tvc : float
    """

    import numpy as np
    import pandas as pd

    df = pd.read_csv(filepath, header=None)

    bin_cols = list(range(36))
    depth_col = 40
    volume_col = 50

    # --- Handle row selection ---
    if isinstance(row_selection, (list, tuple)) and len(row_selection) == 2:
        profile = df.iloc[row_selection[0]:row_selection[1] + 1]
    else:
        profile = df.iloc[row_selection]

    # --- Reference depth ---
    depth0 = profile[depth_col].min()

    histograms = []
    tvc_values = []

    for _, row in profile.iterrows():

        depth = row[depth_col] - depth0

        if depth_range is not None:
            if not (depth_range[0] <= depth <= depth_range[1]):
                continue

        H = row[bin_cols].values.astype(float)
        H = np.nan_to_num(H, nan=0.0)

        V = pd.to_numeric(row[volume_col], errors='coerce')

        if normalize:
            total = np.sum(H)
            if total > 0:
                H = H / total

        histograms.append(H)

        if np.isfinite(V):
            tvc_values.append(V)

    if len(histograms) == 0:
        raise ValueError("No valid histograms found for given selection")

    histograms = np.array(histograms)

    # --- Mean histogram ---
    mean_hist = np.nanmean(histograms, axis=0)

    # --- TVC stats ---
    tvc_values = np.array(tvc_values)

    mean_tvc = np.nanmean(tvc_values) if len(tvc_values) > 0 else np.nan
    std_tvc = np.nanstd(tvc_values) if len(tvc_values) > 0 else np.nan

    
    # --- Plot ---
    if plot:
        import matplotlib.pyplot as plt

        bin_centers = np.logspace(np.log10(1), np.log10(500), len(mean_hist))

        plt.figure(figsize=(4,4))
        # Compute bin edges properly
        edges = [1.00, 1.48, 1.74, 2.05, 2.42, 2.86, 3.38, 3.98, 4.70,
                 5.55, 6.55, 7.72, 9.12, 10.8, 12.7, 15.0, 17.7,
                 20.9, 24.6, 29.1, 34.3, 40.5, 47.7, 56.3, 66.5,
                 78.4, 92.6, 109, 129, 152, 180, 212, 250,
                 297, 354, 420, 500]

    
        log_edges = np.log10(edges)

        # Compute uniform log bin width
        dlog = np.mean(np.diff(log_edges))

        spacing_factor = 0.9

        # Apply uniform width to ALL bins
        new_widths_log = np.full(len(mean_hist), dlog * spacing_factor)

        # Center bins properly
        left_edges_log = log_edges[:-1] + (dlog * (1 - spacing_factor)) / 2

        # Convert back to linear space
        left_edges = 10**left_edges_log
        new_widths = 10**(left_edges_log + new_widths_log) - 10**left_edges_log

        plt.bar(left_edges, mean_hist,
                width=new_widths,
                align='edge',
                color='white',
                edgecolor='black',
                hatch='///')

        plt.xscale('log')

        
        

        tick_values = [1, 5, 25, 100, 500]
        plt.xticks(tick_values, labels=[str(v) for v in tick_values])

        plt.xlabel("Particle Size (μm)")
        plt.ylabel("Mean Volume Concentration (μL/L)")

        
        if normalize:
            title += " (Normalized)"
        plt.title(title)

        plt.grid(True, which='major', alpha=0.3, linewidth=0.5)
        plt.tight_layout()
        plt.show()

    return mean_hist, mean_tvc, std_tvc

def biovolume(filepath, array, title):
    '''
    return average TVC and CHL signal based on sample windows. 
    Arrays are input as [startpt, endpt, startpt, endpt, ...]
    Convert average CHL signal to ug/L chl with formula y = 30.461x
    '''
    df = pd.read_csv(filepath, header=None)
    
    TVC_col = 50
    CHL_col = 58

    segments = []

    # build segments
    for i in range(0, len(array), 2):
        start = array[i]
        end = array[i + 1]
        segment = df.iloc[start:end]
        segments.append(segment)
    
    # combine all segments
    filtered_df = pd.concat(segments)

    # build paired dataframe (keeps TVC + CHL aligned)
    paired = pd.DataFrame({
        "TVC": pd.to_numeric(filtered_df.iloc[:, TVC_col], errors='coerce'),
        "CHL": pd.to_numeric(filtered_df.iloc[:, CHL_col], errors='coerce')
    })

    # drop NaNs
    paired = paired.dropna()

    # compute 95th percentile thresholds
    TVC_thresh = paired["TVC"].quantile(0.95)
    CHL_thresh = paired["CHL"].quantile(0.95)

    # remove rows where either value exceeds its threshold
    paired_filtered = paired[
        (paired["TVC"] <= TVC_thresh) &
        (paired["CHL"] <= CHL_thresh)
    ]

    # compute averages
    avg_TVC = paired_filtered["TVC"].mean()
    avg_CHL_signal = paired_filtered["CHL"].mean()

    # convert CHL to ug/L
    chl_ugL = 30.461 * avg_CHL_signal

    
    print("Avg TVC (ppm): ", avg_TVC)
    print("Avg CHL (V): ", avg_CHL_signal)
    print("Avg CHL (ug/L): ", chl_ugL)

    return avg_TVC, avg_CHL_signal, chl_ugL


# ----- Arrays: full depth -------
l0_full, l0_95_full, l0_tvc_full, l0_sd_full = mean_histogram_shape("lap_baseline_rs.csv", lap_baseline)
lt3_full, lt3_95, lt3_TVC_full, lt3_tvc_sd_full = mean_histogram_shape("lap_t3_rs.csv", lap_t3, condition="Opt")
lt23_full, lt23_95, lt23_TVC_full, lt23_tvc_sd_full = mean_histogram_shape("lap_t23_rs.csv", lap_t23, condition="Opt") 

k0_full, k0_95_full, k0_tvc_full, k0_sd_full = mean_histogram_shape("kao_baseline_rs.csv", kao_baseline)
kt3_full, kt3_95, kt3_TVC_full, kt3_tvc_sd_full = mean_histogram_shape("kao_t22_rs.csv", kao_t3, condition="Opt")
kt22_full, kt22_95, kt22_TVC_full, kt22_tvc_sd_full = mean_histogram_shape("kao_t22_rs.csv", kao_t22, condition="Opt") 



# ----- Arrays: 10 to 20 cm -------
lt0_slice, lt0_tvc_slice, kt0_slice, kt0_tvc_slice = specific_depth_averaged_histograms( filepath_t1="lap_baseline_rs.csv", filepath_t2="kao_baseline_rs.csv", sample_indices_t1=lap_baseline, sample_indices_t2=kao_baseline, 
                                        depth_range=(0.1, 0.2), clip_percentile=None, condition='Opt', vmax=None)
kt3_slice, kt3_tvc_slice, kt22_slice, kt22_tvc_slice = specific_depth_averaged_histograms( filepath_t1="kao_t3_rs.csv", filepath_t2="kao_t22_rs.csv", sample_indices_t1=kao_t3, sample_indices_t2=kao_t22, 
                                        depth_range=(0.1, 0.2), clip_percentile=None, condition='Opt', vmax=None)
lt3_slice, lt3_tvc_slice, lt23_slice, lt23_tvc_slice = specific_depth_averaged_histograms( filepath_t1="lap_t3_rs.csv", filepath_t2="lap_t23_rs.csv", sample_indices_t1=lap_t3, sample_indices_t2=lap_t23, 
                                        depth_range=(0.1, 0.2), clip_percentile=None, condition='Opt', vmax=None)

kt3_hist, kt3_error, kt3_tvc, kt3_tvcsd  = mean_histogram_shape("kao_t3_rs.csv", kao_t3, depth_range=(0.1, 0.2), condition="Opt", title = "Kaolinite Optimal Dose: T = 3 hours")
kt22_hist, kt22_error, kt22_tvc, kt22_tvcsd = mean_histogram_shape("kao_t22_rs.csv", kao_t22, depth_range=(0.1, 0.2), condition="Opt", title = "Kaolinite Optimal Dose: T = 22 hours")
lt3_hist, lt3_error, lt3_tvc, lt3_tvcsd = mean_histogram_shape("lap_t3_rs.csv", lap_t3, depth_range=(0.1, 0.2), condition="Opt", title = "Laponite Optimal Dose: T = 3 hours")
lt23_hist, lt23_error, lt23_tvc, lt23_tvcsd = mean_histogram_shape("lap_t23_rs.csv", lap_t23, depth_range=(0.1, 0.2), condition="Opt", title = "Laponite Optimal Dose: T = 23 hours")


# ---- table 1 statistics ----
print('---- Table 1 ----')
print('Background t0: ', summarize_histogram(l0_full, l0_tvc_full))

kao_bg = [25]
mean_hist = mean_histogram_shape_simple("kaolinite_background_4gL_rs.csv", kao_bg)
print('Kaolinite t0: ', summarize_histogram(hist=mean_hist[0], tvc=mean_hist[1]))

lap_bg = [2, 3, 4, 5, 6, 7, 8, 9]
mean_hist = mean_histogram_shape_simple("laponite_background_0-05gL_rs.csv", lap_bg)
print('Laponite t0: ', summarize_histogram(hist=mean_hist[0], tvc=mean_hist[1]))

# ---- table 2 statistics ----
print('---- Table 2 ----')
print('Kaolinite t3: ', summarize_histogram(kt3_slice, kt3_tvc_slice))
print('Kaolinite t22: ', summarize_histogram(kt22_slice, kt22_tvc_slice))
print('Laponite t3: ', summarize_histogram(lt3_slice, lt3_tvc_slice))
print('Laponite t23: ', summarize_histogram(lt23_slice, lt23_tvc_slice))



# ---- biovolume via chlorophyll
def biovolume(filepath, array, claytype='L'):
    '''
    return average TVC and CHL signal based on sample windows. 
    Arrays are input as [startpt, endpt, startpt, endpt, ...]
    Convert average CHL signal to ug/L chl with formula y = 30.461x
    '''
    df = pd.read_csv(filepath, header=None)
    
    TVC_col = 50
    CHL_col = 58

    segments = []

    # build segments
    for i in range(0, len(array), 2):
        start = array[i]
        end = array[i + 1]
        segment = df.iloc[start:end]
        segments.append(segment)
    
    # combine all segments
    filtered_df = pd.concat(segments)

    # build paired dataframe (keeps TVC + CHL aligned)
    paired = pd.DataFrame({
        "TVC": pd.to_numeric(filtered_df.iloc[:, TVC_col], errors='coerce'),
        "CHL": pd.to_numeric(filtered_df.iloc[:, CHL_col], errors='coerce')
    })

    # drop NaNs
    paired = paired.dropna()

    # compute 95th percentile thresholds
    TVC_thresh = paired["TVC"].quantile(0.95)
    CHL_thresh = paired["CHL"].quantile(0.95)

    # remove rows where either value exceeds its threshold
    paired_filtered = paired[
        (paired["TVC"] <= TVC_thresh) &
        (paired["CHL"] <= CHL_thresh)
    ]

    # compute averages
    avg_TVC = paired_filtered["TVC"].mean()
    avg_CHL_signal = paired_filtered["CHL"].mean()

    # convert CHL to ug/L
    if claytype == 'L':
        chl_ugL = 28.13 * avg_CHL_signal + 0.09
    elif claytype == 'K':
        chl_ugL = 3.05 * avg_CHL_signal + 0.717

    
    print("Avg TVC (ppm): ", avg_TVC)
    print("Avg CHL (V): ", avg_CHL_signal)
    print("Avg CHL (ug/L): ", chl_ugL)

    return avg_TVC, avg_CHL_signal, chl_ugL

L_BL_t3 = [62, 86, 125, 150, 240, 269, 301, 325]
L_opt_t3 = [876, 911, 939, 969, 1003, 1036, 1065, 1096]
L_opt_t23 = [655, 678, 709, 731, 767, 788, 814, 835]

print('---Laponite BL---')
biovolume(filepath='lap_t3_rs.csv', array=L_BL_t3, claytype='L')
print('---Laponite t3---')
biovolume(filepath='lap_t3_rs.csv', array=L_opt_t3, claytype='L')
print('---Laponite t23---')
biovolume(filepath='lap_t23_rs.csv', array=L_opt_t23, claytype='L')

K_BL_t22 = [89, 127, 153, 197, 347, 383, 494, 521]
K_opt_t3 = [1808, 1833, 1868, 1894, 1932, 1957, 2077, 2103]
K_opt_t22 = [1487, 1496, 1623, 1636, 1759, 1774, 1954, 1977]

print('---Kaolinite BL---')
biovolume(filepath='kao_t22_rs.csv', array=K_BL_t22, claytype='K')
print('---Kaolinite t3---')
biovolume(filepath='kao_t3_rs.csv', array=K_opt_t3, claytype='K')
print('---Kaolinite t22---')
biovolume(filepath='kao_t22_rs.csv', array=K_opt_t22, claytype='K')
```

```python
# Figure 2
def mean_histogram_shape_simple(
    filepath,
    row_selection,
    depth_range=None,
    normalize=False,
    plot=True, title=''
):
    """
    Compute mean histogram shape from selected rows.

    Parameters
    ----------
    filepath : str
        CSV file path

    row_selection : list or tuple
        Either:
        - [start, end] for a continuous range
        - list of row indices

    depth_range : tuple, optional
        Restrict averaging to (min_depth, max_depth)

    normalize : bool
        If True, normalize each histogram before averaging

    plot : bool
        Whether to plot the result

    Returns
    -------
    mean_hist : ndarray
    mean_tvc : float
    std_tvc : float
    """

    df = pd.read_csv(filepath, header=None)

    bin_cols = list(range(36))
    depth_col = 40
    volume_col = 50

    # --- Handle row selection ---
    if isinstance(row_selection, (list, tuple)) and len(row_selection) == 2:
        profile = df.iloc[row_selection[0]:row_selection[1] + 1]
    else:
        profile = df.iloc[row_selection]

    # --- Reference depth ---
    depth0 = profile[depth_col].min()

    histograms = []
    tvc_values = []

    for _, row in profile.iterrows():

        depth = row[depth_col] - depth0

        if depth_range is not None:
            if not (depth_range[0] <= depth <= depth_range[1]):
                continue

        H = row[bin_cols].values.astype(float)
        H = np.nan_to_num(H, nan=0.0)

        V = pd.to_numeric(row[volume_col], errors='coerce')

        if normalize:
            total = np.sum(H)
            if total > 0:
                H = H / total

        histograms.append(H)

        if np.isfinite(V):
            tvc_values.append(V)

    if len(histograms) == 0:
        raise ValueError("No valid histograms found for given selection")

    histograms = np.array(histograms)

    # --- Mean histogram ---
    mean_hist = np.nanmean(histograms, axis=0)

    # --- TVC stats ---
    tvc_values = np.array(tvc_values)

    mean_tvc = np.nanmean(tvc_values) if len(tvc_values) > 0 else np.nan
    std_tvc = np.nanstd(tvc_values) if len(tvc_values) > 0 else np.nan

    
    # --- Plot ---
    if plot:

        plt.figure(figsize=(4,4.1))
        # Compute bin edges properly
        edges = [1.00, 1.48, 1.74, 2.05, 2.42, 2.86, 3.38, 3.98, 4.70,
                 5.55, 6.55, 7.72, 9.12, 10.8, 12.7, 15.0, 17.7,
                 20.9, 24.6, 29.1, 34.3, 40.5, 47.7, 56.3, 66.5,
                 78.4, 92.6, 109, 129, 152, 180, 212, 250,
                 297, 354, 420, 500]

    
        log_edges = np.log10(edges)

        # Compute uniform log bin width
        dlog = np.mean(np.diff(log_edges))

        spacing_factor = 0.9

        # Apply uniform width to ALL bins
        new_widths_log = np.full(len(mean_hist), dlog * spacing_factor)

        # Center bins properly
        left_edges_log = log_edges[:-1] + (dlog * (1 - spacing_factor)) / 2

        # Convert back to linear space
        left_edges = 10**left_edges_log
        new_widths = 10**(left_edges_log + new_widths_log) - 10**left_edges_log

        plt.bar(left_edges, mean_hist,
                width=new_widths,
                align='edge',
                color='red',
                edgecolor='black',
                hatch='///')

        plt.xscale('log')

        
        

        tick_values = [1, 5, 25, 100, 500]
        plt.xticks(tick_values, labels=[str(v) for v in tick_values])

        plt.xlabel("Particle Size (μm)", fontsize=15)
        plt.ylabel("Mean Volume Concentration (μL/L)", fontsize=15)

        
        if normalize:
            title += " (Normalized)"
        plt.title(title, fontsize=15)

        plt.grid(True, which='major', alpha=0.3, linewidth=0.5)
        plt.tight_layout()
        plt.show()

    return mean_hist, mean_tvc, std_tvc

kao_bg = [25]
mean_hist = mean_histogram_shape_simple("kaolinite_background_4gL_rs.csv", kao_bg, title = "(c) 4 g/L Kaolinite")
summarize_histogram(hist=mean_hist[0], tvc=mean_hist[1])

lap_bg = [2, 3, 4, 5, 6, 7, 8, 9]
mean_hist = mean_histogram_shape_simple("laponite_background_0-05gL_rs.csv", lap_bg, title = "(b) 0.05 g/L Laponite")
summarize_histogram(hist=mean_hist[0], tvc=mean_hist[1])
```

```python
# Figure 3
def plot_depth_averaged_histograms_two_times(
    filepath_t1,
    filepath_t2,
    sample_indices_t1,
    sample_indices_t2,
    depth_range=None,
    clip_percentile=None,
    condition=None,
    vmax=None,
    title=None
):
    """
    Compare histogram-integrated volumes vs depth for two timepoints.

    Parameters
    ----------
    filepath_t1, filepath_t2 : str
        CSV file paths

    sample_indices_t1, sample_indices_t2 : list
        Sampling windows

    depth_range : tuple, optional
        Restrict depth region (min, max)

    clip_percentile : float, optional
        Remove outliers (e.g., 95 keeps middle 95%)

    condition : str or list, optional
        Select subset of conditions

    vmax : float, optional
        Fix x-axis max
    """

    df1 = pd.read_csv(filepath_t1, header=None)
    df2 = pd.read_csv(filepath_t2, header=None)

    
    depth_col = 40

    group_labels = ["Baseline", "Low", "Med", "High", "Mix"]

    # --- Condition filter ---
    if condition is not None:
        if isinstance(condition, str):
            condition = [condition]
        group_labels = [c for c in group_labels if c in condition]

    tanks_per_condition = 3
    profiles_per_tank = 2

    def build_condition_runs(sample_indices):
        runs = list(zip(sample_indices[0::2], sample_indices[1::2]))
        cond_dict = {}
        for i, label in enumerate(["Baseline", "Low", "Med", "High", "Mix"]):
            start = i * tanks_per_condition * profiles_per_tank
            end = start + tanks_per_condition * profiles_per_tank
            cond_dict[label] = runs[start:end]
        return cond_dict

    cond_runs_t1 = build_condition_runs(sample_indices_t1)
    cond_runs_t2 = build_condition_runs(sample_indices_t2)

    def compute_profiles(df, cond_runs):

        results = {}

        for label in group_labels:

            runs = cond_runs[label]
            depths_all = []
            volumes_all = []

            for run_start, run_end in runs:

                profile = df.iloc[run_start:run_end + 1]
                depth0 = profile[depth_col].min()

                volume_col = 50 

                for _, row in profile.iterrows():

                    depth = row[depth_col] - depth0

                    if depth_range is not None:
                        if not (depth_range[0] <= depth <= depth_range[1]):
                            continue

                    V = pd.to_numeric(row[volume_col], errors='coerce')

                    if not np.isnan(V):
                        depths_all.append(depth)
                        volumes_all.append(V)

            depths_all = np.array(depths_all)
            volumes_all = np.array(volumes_all)

            # --- Clip outliers ---
            if clip_percentile is not None and len(volumes_all) > 0:
                lo, hi = np.percentile(
                    volumes_all,
                    [(100 - clip_percentile)/2, 100 - (100 - clip_percentile)/2]
                )
                mask = (volumes_all >= lo) & (volumes_all <= hi)
                depths_all = depths_all[mask]
                volumes_all = volumes_all[mask]

            results[label] = (depths_all, volumes_all)

        return results

    results_t1 = compute_profiles(df1, cond_runs_t1)
    results_t2 = compute_profiles(df2, cond_runs_t2)

    # --- Styling ---
    from matplotlib.lines import Line2D
    plt.rcParams.update({
        "font.size": 12,
        "axes.spines.top": False,
        "axes.spines.right": False
    })

    fig, axes = plt.subplots(
        1, len(group_labels),
        figsize=(5*len(group_labels), 5.5),
        sharey=True
    )

    if len(group_labels) == 1:
        axes = [axes]

    if vmax is None:
        vmax_list = [None] * len(group_labels)

    elif isinstance(vmax, (list, tuple, np.ndarray)):
        if len(vmax) != len(group_labels):
            raise ValueError("Length of vmax list must match number of conditions")
        vmax_list = vmax

    else:
        vmax_list = [vmax] * len(group_labels)

    for ax, label, vmax_i in zip(axes, group_labels, vmax_list):

        d1, v1 = results_t1[label]
        d2, v2 = results_t2[label]

        ax.scatter(
            v1, d1,
            facecolors="white",
            edgecolors="red",
            alpha=1,
            s=55,
            linewidth=1.5
        )

        ax.scatter(
            v2, d2,
            facecolors="red",
            edgecolors="black",
            marker="s",
            alpha=0.25,
            s=55,
            linewidth=1.5
        )

        # --- Independent means per subplot ---
        avg1 = np.nanmean(v1) if len(v1) > 0 else np.nan
        avg2 = np.nanmean(v2) if len(v2) > 0 else np.nan

        # Draw vertical lines
        if np.isfinite(avg1):
            ax.axvline(avg1, color="black", linestyle=":", linewidth=2.5)

        if np.isfinite(avg2):
            ax.axvline(avg2, color="black", linestyle="-", linewidth=2.5)

        # --- Annotation box ---
        label1 = r"$\overline{\mathrm{TVC}}_3$"
        label2 = r"$\overline{\mathrm{TVC}}_{22}$"

        textstr = (
            f"{label1} = {avg1:.1f}\n"
            f"{label2} = {avg2:.1f}"
        )

        ax.text(
            0.97, 0.97,
            textstr,
            transform=ax.transAxes,
            fontsize=14,
            verticalalignment="top",
            horizontalalignment="right",
            bbox=dict(
                boxstyle="round,pad=0.3",
                facecolor="white",
                edgecolor="0.6",
                alpha=0.9
            )
        )

        ax.set_title(label, fontsize=16)
        ax.invert_yaxis()
        ax.set_ylim(0.55, 0)
        ax.grid(True, alpha=0.25)

        if vmax_i is not None:
            ax.set_xlim(0, vmax_i)


    axes[0].set_ylabel("Normalized Depth [cm]", fontsize=16)
    fig.supxlabel("Total Volume Concentration [ppm]", y=0.05, fontsize=16)

    legend_elements = [
        Line2D(
            [0], [0],
            color="black",
            linestyle=":",
            marker="o",
            markerfacecolor="white",
            markeredgecolor="red",
            markersize=7,
            linewidth=2,
            label="Time: 3 hours"
            ),
        Line2D(
            [0], [0],
            color="black",
            linestyle="-",
            marker="s",
            markerfacecolor="red",
            alpha = 0.5,
            markersize=7,
            linewidth=2,
            label="Time: 22 hours"
        )
    ]

    fig.legend(
        handles=legend_elements,
        loc="upper right",
        ncol=2,
        frameon=False,
        bbox_to_anchor=(0.5, 1.02), fontsize=16
    )

    fig.text(
        0.98, 1,
        r"$\overline{\mathrm{TVC}}_t$: Depth-averaged Total Volume Concentration for each sampling session $\it{t}$",
        ha='right', va='top',
        fontsize=16,
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
    )

    fig.text(
        0.075, 1,
        "(a) Kaolinite",
        ha='right', va='top',
        fontsize=16,
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
    )


    if title is not None:
        fig.suptitle(title, fontsize=14, y=1.08)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

plot_depth_averaged_histograms_two_times(
    "kao_t3_rs.csv",
    "kao_t22_rs.csv",
    kao_t3,
    kao_t22, 
    vmax=[50, 10000, 10000, 10000], clip_percentile=99, condition=["Baseline", "Low", "Med", "High"],
    title=None
)

plot_depth_averaged_histograms_two_times(
    "lap_t3_rs.csv",
    "lap_t23_rs.csv",
    lap_t3,
    lap_t23, 
    vmax=[50, 50, 300, 10000], clip_percentile=99, condition=["Baseline", "Low", "Med", "High"],
    title=None
)
```

```python
# Figure 4
def plot_depth_averaged_histogram_d50_two_times(
    filepath_t1,
    filepath_t2,
    sample_indices_t1,
    sample_indices_t2,
    depth_range=None,
    clip_percentile=None,
    condition=None,
    vmax=None
):
    """
    Compare D50 (median particle size) vs depth for two timepoints.
    """

    df1 = pd.read_csv(filepath_t1, header=None)
    df2 = pd.read_csv(filepath_t2, header=None)

    bin_cols = list(range(36))
    depth_col = 40

    bin_widths = np.array([
        0.48, 0.26, 0.31, 0.37, 0.44, 0.52, 0.60, 0.72, 0.85, 1,
        1.17, 1.4, 1.68, 1.9, 2.3, 2.7, 3.2, 3.7, 4.5, 5.2,
        6.2, 7.2, 8.6, 10.2, 11.9, 14.2, 16.4, 20, 23, 28,
        32, 38, 47, 57, 66, 80
    ])

    # --- Bin centers ---
    bin_centers = np.cumsum(bin_widths) - bin_widths / 2
    log_centers = np.log10(bin_centers)

    group_labels = ["Baseline", "Low", "Med", "High", "Mix"]

    if condition is not None:
        if isinstance(condition, str):
            condition = [condition]
        group_labels = [c for c in group_labels if c in condition]

    tanks_per_condition = 3
    profiles_per_tank = 2

    def build_condition_runs(sample_indices):
        runs = list(zip(sample_indices[0::2], sample_indices[1::2]))
        cond_dict = {}
        for i, label in enumerate(["Baseline", "Low", "Med", "High", "Mix"]):
            start = i * tanks_per_condition * profiles_per_tank
            end = start + tanks_per_condition * profiles_per_tank
            cond_dict[label] = runs[start:end]
        return cond_dict

    cond_runs_t1 = build_condition_runs(sample_indices_t1)
    cond_runs_t2 = build_condition_runs(sample_indices_t2)

    def compute_profiles(df, cond_runs):

        results = {}

        for label in group_labels:

            runs = cond_runs[label]
            depths_all = []
            d50_all = []

            for run_start, run_end in runs:

                profile = df.iloc[run_start:run_end + 1]
                depth0 = profile[depth_col].min()

                for _, row in profile.iterrows():

                    bin_heights = row[bin_cols].values.astype(float)
                    bin_heights = np.nan_to_num(bin_heights, nan=0.0)

                    total_mass = np.sum(bin_heights)
                    if total_mass == 0:
                        continue

                    # --- PDF + CDF ---
                    pdf = bin_heights / total_mass
                    cdf = np.cumsum(pdf)

                    # --- D50 (log-space interpolation) ---
                    D50 = 10**np.interp(0.5, cdf, log_centers)

                    depth = row[depth_col] - depth0

                    if depth_range is not None:
                        if not (depth_range[0] <= depth <= depth_range[1]):
                            continue

                    if np.isfinite(D50):
                        depths_all.append(depth)
                        d50_all.append(D50)

            depths_all = np.array(depths_all)
            d50_all = np.array(d50_all)

            # --- Clip outliers ---
            if clip_percentile is not None and len(d50_all) > 0:
                lo, hi = np.percentile(
                    d50_all,
                    [(100 - clip_percentile)/2, 100 - (100 - clip_percentile)/2]
                )
                mask = (d50_all >= lo) & (d50_all <= hi)
                depths_all = depths_all[mask]
                d50_all = d50_all[mask]

            results[label] = (depths_all, d50_all)

        return results

    results_t1 = compute_profiles(df1, cond_runs_t1)
    results_t2 = compute_profiles(df2, cond_runs_t2)

    # --- Plot styling ---
    from matplotlib.lines import Line2D
    plt.rcParams.update({
        "font.size": 12,
        "axes.spines.top": False,
        "axes.spines.right": False
    })

    fig, axes = plt.subplots(
        1, len(group_labels),
        figsize=(5*len(group_labels), 5.5),
        sharey=True
    )

    if len(group_labels) == 1:
        axes = [axes]

    if vmax is None:
        vmax_list = [None] * len(group_labels)
    elif isinstance(vmax, (list, tuple, np.ndarray)):
        if len(vmax) != len(group_labels):
            raise ValueError("Length of vmax list must match number of conditions")
        vmax_list = vmax
    else:
        vmax_list = [vmax] * len(group_labels)

    for ax, label, vmax_i in zip(axes, group_labels, vmax_list):

        d1, d50_1 = results_t1[label]
        d2, d50_2 = results_t2[label]

        # --- Scatter ---
        ax.scatter(
            d50_1, d1,
            facecolors="white",
            edgecolors="red",
            alpha=1,
            s=55,
            linewidth=1.5
        )

        ax.scatter(
            d50_2, d2,
            facecolors="red",
            edgecolors="black",
            marker="s",
            alpha=0.25,
            s=55,
            linewidth=1.5
        )

        # --- Means ---
        avg1 = np.nanmean(d50_1) if len(d50_1) > 0 else np.nan
        avg2 = np.nanmean(d50_2) if len(d50_2) > 0 else np.nan

        if np.isfinite(avg1):
            ax.axvline(avg1, color="black", linestyle=":", linewidth=2.5)

        if np.isfinite(avg2):
            ax.axvline(avg2, color="black", linestyle="-", linewidth=2.5)

        # --- Annotation ---

        label1 = r"$\overline{\mathrm{D50}}_3$"
        label2 = r"$\overline{\mathrm{D50}}_{22}$"

        textstr = (
            f"{label1} = {avg1:.1f} µm\n"
            f"{label2} = {avg2:.1f} µm"
        )

        ax.text(
            0.97, 0.97,
            textstr, fontsize=16,
            transform=ax.transAxes,
            verticalalignment="top",
            horizontalalignment="right",
            bbox=dict(facecolor="white", alpha=0.9)
        )

        ax.set_title(label, fontsize=16)
        ax.invert_yaxis()
        ax.set_ylim(0.55, 0)
        ax.grid(True, alpha=0.25)

        if vmax_i is not None:
            ax.set_xlim(0, vmax_i)

    axes[0].set_ylabel("Depth [cm]", fontsize=16)
    fig.supxlabel("D50 [µm]", y=0.05, fontsize=16)

    legend_elements = [
        Line2D(
            [0], [0],
            color="black",
            linestyle=":",
            marker="o",
            markerfacecolor="white",
            markeredgecolor="red",
            markersize=7,
            linewidth=2,
            label="Time: 3 hours"
        ),
        Line2D(
            [0], [0],
            color="black",
            linestyle="-",
            marker="s",
            markerfacecolor="red",
            alpha=0.5,
            markeredgecolor="black",
            markersize=7,
            linewidth=2,
            label="Time: 22 hours"
        )
    ]

    fig.legend(
        handles=legend_elements,
        loc="upper right",
        ncol=2,
        frameon=False,
        bbox_to_anchor=(0.5, 1.02), fontsize=16
    )


    fig.text(
        0.98, 0.98,
        r"$\overline{\mathrm{D50}}_t$: Depth-averaged median diameter D50 for each sampling session $\it{t}$",
        ha='right', va='top',
        fontsize=16,
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
    )
    fig.text(
        0.075, 1,
        "(a) Kaolinite",
        ha='right', va='top',
        fontsize=16,
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
    )
    
    title=None
    fig.suptitle(title, fontsize=16, y=1.08)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

plot_depth_averaged_histogram_d50_two_times("kao_t3_rs.csv", "kao_t22_rs.csv", kao_t3, kao_t22, vmax=500, clip_percentile=99, condition=["Baseline", "Low", "Med", "High"])
plot_depth_averaged_histogram_d50_two_times("lap_t3_rs.csv", "lap_t23_rs.csv", lap_t3, lap_t23, vmax=500, clip_percentile=99, condition=["Baseline", "Low", "Med", "High"])
```

```python
# Figure 5
def plotk(filepath_t1=None, filepath_t2=None, filepath_t3=None, 
          sample_window1=None, sample_window2=None, sample_window3=None, 
          rejection_threshold=0.6, title=None, time_labels=[3, 23, 27],
          depth_range = (0.1, 0.2), condition="Opt", clip_percentile=None, bin_id=None, reaction_order='One'):
    
    # read df
    df1 = pd.read_csv(filepath_t1, header=None)
    df2 = pd.read_csv(filepath_t2, header=None)
    df3 = pd.read_csv(filepath_t3, header=None)
    TVC_col = 50
    depth_col = 40

    bin_cols = list(range(36))
    depth_col = 40

    group_labels = ["Baseline", "Low", "Med", "High", "Mix"]

    # --- Condition filter ---
    if condition is not None:
        if isinstance(condition, str):
            condition = [condition]
        group_labels = [c for c in group_labels if c in condition]

    tanks_per_condition = 3
    profiles_per_tank = 2

    def build_condition_runs(sample_indices):
        runs = list(zip(sample_indices[0::2], sample_indices[1::2]))
        cond_dict = {}
        for i, label in enumerate(["Baseline", "Low", "Med", "High", "Mix"]):
            start = i * tanks_per_condition * profiles_per_tank
            end = start + tanks_per_condition * profiles_per_tank
            cond_dict[label] = runs[start:end]
        return cond_dict

    cond_runs_t1 = build_condition_runs(sample_window1)
    cond_runs_t2 = build_condition_runs(sample_window2)
    cond_runs_t3 = build_condition_runs(sample_window3)

    def compute_profiles(df, cond_runs):

        results = {}

        for label in group_labels:

            runs = cond_runs[label]
            depths_all = []
            volumes_all = []

            for run_start, run_end in runs:

                profile = df.iloc[run_start:run_end + 1]
                depth0 = profile[depth_col].min()

                for _, row in profile.iterrows():
                    depth = row[depth_col] - depth0
                    # extract TVC for each sample in the depth region
                    TVC = row[bin_cols].values.astype(float)

                    if depth_range is not None:
                        if not (depth_range[0] <= depth <= depth_range[1]):
                            continue

                    if np.all(np.isfinite(TVC)):
                        depths_all.append(depth)
                        volumes_all.append(TVC)

            depths_all = np.array(depths_all)
            volumes_all = np.array(volumes_all)

            # --- Clip outliers ---
            if clip_percentile is not None and len(volumes_all) > 0:
                lo, hi = np.percentile(
                    volumes_all,
                    [(100 - clip_percentile)/2, 100 - (100 - clip_percentile)/2]
                )
                mask = (volumes_all >= lo) & (volumes_all <= hi)
                depths_all = depths_all[mask]
                volumes_all = volumes_all[mask]

            results[label] = (depths_all, volumes_all)

        return results
    
    def plot_bin_lnC_vs_time(results_t1, results_t2, results_t3,
                          bin_index,
                          time_labels=[3, 23, 27],
                          condition_labels=None,
                          rejection_threshold=None,
                          title=None):
        """
        Plot ln(C) vs time for a specific bin (1–36) 1st order reaction.
        """

        bin_i = bin_index - 1
        times = np.array(time_labels)

        def extract_C(results):
            """Average concentration for a given bin across all conditions"""
            C_vals = []

            for label in results:
                depths, volumes = results[label]

                if len(volumes) == 0:
                    continue

                volumes = np.array(volumes)

                if volumes.ndim != 2:
                    continue

                bin_vals = volumes[:, bin_i]

                if len(bin_vals) > 0:
                    C_vals.append(np.nanmean(bin_vals))

            return np.nanmean(C_vals) if len(C_vals) > 0 else np.nan

        C_t1 = extract_C(results_t1)
        C_t2 = extract_C(results_t2)
        C_t3 = extract_C(results_t3)

        C_vals = np.array([C_t1, C_t2, C_t3])

        # remove invalid
        if np.any(~np.isfinite(C_vals)) or np.any(C_vals <= 0):
            print("Invalid data for this bin (NaN or zero concentrations).")
            return

        lnC = np.log(C_vals + 1e-12)

        # linear fit
        slope, intercept = np.polyfit(times, lnC, 1)
        fit = slope * times + intercept

        # R^2
        ss_res = np.sum((lnC - fit)**2)
        ss_tot = np.sum((lnC - np.mean(lnC))**2)
        r2 = 1 - ss_res / ss_tot if ss_tot > 0 else 0

        # ---- Plot ----
        plt.figure(figsize=(6,4))
        plt.scatter(times, lnC, color='white', edgecolor='black')
        plt.plot(times, fit, linestyle="--", label=f"k={slope:.3f} [1/h], R²={r2:.3f}", color='black')

        plt.xlabel("Time (hrs)")
        plt.ylabel("ln(C)")
        plt.title(title if title else f"Bin {bin_index}: ln(C) vs time")
        plt.grid(True)
        plt.legend()
        plt.show()

        print(f"Bin {bin_index} results:")
        print(f"  C(t): {C_vals}")
        print(f"  slope = {slope:.4f}")
        print(f"  k = {-slope:.4f} [1/h]")
        print(f"  R² = {r2:.4f}")

    def plot_bin_C_vs_time(results_t1, results_t2, results_t3,
                          bin_index,
                          time_labels=[3, 23, 27],
                          condition_labels=None,
                          rejection_threshold=None,
                          title=None):
        """
        Plot C vs time for a specific bin (1–36) 0th order reaction.
        """

        bin_i = bin_index - 1
        times = np.array(time_labels)

        def extract_C(results):
            """Average concentration for a given bin across all conditions"""
            C_vals = []

            for label in results:
                depths, volumes = results[label]

                if len(volumes) == 0:
                    continue

                volumes = np.array(volumes)

                if volumes.ndim != 2:
                    continue

                bin_vals = volumes[:, bin_i]

                if len(bin_vals) > 0:
                    C_vals.append(np.nanmean(bin_vals))

            return np.nanmean(C_vals) if len(C_vals) > 0 else np.nan

        C_t1 = extract_C(results_t1)
        C_t2 = extract_C(results_t2)
        C_t3 = extract_C(results_t3)

        C_vals = np.array([C_t1, C_t2, C_t3])

        # remove invalid
        if np.any(~np.isfinite(C_vals)) or np.any(C_vals <= 0):
            print("Invalid data for this bin (NaN or zero concentrations).")
            return

        # linear fit
        slope, intercept = np.polyfit(times, C_vals, 1)
        fit = slope * times + intercept

        # R^2
        ss_res = np.sum((C_vals - fit)**2)
        ss_tot = np.sum((C_vals - np.mean(C_vals))**2)
        r2 = 1 - ss_res / ss_tot if ss_tot > 0 else 0

        # ---- Plot ----
        plt.figure(figsize=(6,4))
        plt.scatter(times, C_vals, label="data", color='white', edgecolor='black')
        plt.plot(times, fit, linestyle="--", label=f"k={slope:.3f} [μL/(L \u00B7 h)], R²={r2:.3f}", color='black', markeredgecolor='black', markerfacecolor='white')

        plt.xlabel("Time (hrs)")
        plt.ylabel("C")
        plt.title(title if title else f"Bin {bin_index}: 0th order reaction: C vs time")
        plt.grid(True)
        plt.legend()
        plt.show()

        print(f"Bin {bin_index} results:")
        print(f"  C(t): {C_vals}")
        print(f"  slope = {slope:.4f}")
        print(f"  k = {-slope:.4f} [uL/(L \u00B7 hr)]")
        print(f"  R² = {r2:.4f}")
        return
    
    def plot_bin_1_C_vs_time(results_t1, results_t2, results_t3,
                          bin_index,
                          time_labels=[3, 23, 27],
                          condition_labels=None,
                          rejection_threshold=None,
                          title=None):
        """
        Plot 1/C vs time for a specific bin (1–36) 2nd order reaction.
        """

        bin_i = bin_index - 1
        times = np.array(time_labels)

        def extract_C(results):
            """Average concentration for a given bin across all conditions"""
            C_vals = []

            for label in results:
                depths, volumes = results[label]

                if len(volumes) == 0:
                    continue

                volumes = np.array(volumes)

                if volumes.ndim != 2:
                    continue

                bin_vals = volumes[:, bin_i]

                if len(bin_vals) > 0:
                    C_vals.append(np.nanmean(bin_vals))

            return np.nanmean(C_vals) if len(C_vals) > 0 else np.nan

        C_t1 = extract_C(results_t1)
        C_t2 = extract_C(results_t2)
        C_t3 = extract_C(results_t3)

        C_vals = np.array([C_t1, C_t2, C_t3])

        # remove invalid
        if np.any(~np.isfinite(C_vals)) or np.any(C_vals <= 0):
            print("Invalid data for this bin (NaN or zero concentrations).")
            return

        one_over_C = 1/C_vals

        # linear fit
        slope, intercept = np.polyfit(times, one_over_C, 1)
        fit = slope * times + intercept

        # R^2
        ss_res = np.sum((one_over_C - fit)**2)
        ss_tot = np.sum((one_over_C - np.mean(one_over_C))**2)
        r2 = 1 - ss_res / ss_tot if ss_tot > 0 else 0

        # ---- Plot ----
        plt.figure(figsize=(6,4))
        plt.scatter(times, one_over_C, label="data", color='white', edgecolor='black')
        plt.plot(times, fit, linestyle="--", label=f"k={slope:.3f} [L/(μL \u00B7 h], R²={r2:.3f}", color='black', markeredgecolor='black', markerfacecolor='white')

        plt.xlabel("Time (hrs)")
        plt.ylabel("1/C")
        plt.title(title if title else f"Bin {bin_index}: 1/C vs time")
        plt.grid(True)
        plt.legend()
        plt.show()

        print(f"Bin {bin_index} results:")
        print(f"  C(t): {C_vals}")
        print(f"  slope = {slope:.4f}")
        print(f"  k = {-slope:.4f} [L/uL \u00B7 hr]")
        print(f"  R² = {r2:.4f}")
        
    results_t1 = compute_profiles(df1, cond_runs_t1)
    results_t2 = compute_profiles(df2, cond_runs_t2)
    results_t3 = compute_profiles(df3, cond_runs_t3)

    
    k_dict = {label: [] for label in group_labels}
    times = np.array(time_labels)
    bins = np.arange(36)

    for label in group_labels:

        for b in bins:

            C_vals = []

            for results in [results_t1, results_t2, results_t3]:

                depths, volumes = results[label]

                if len(volumes) == 0:
                    C_vals.append(np.nan)
                    continue

                volumes = np.array(volumes)

                if volumes.ndim != 2:
                    C_vals.append(np.nan)
                    continue

                bin_vals = volumes[:, b]

                if len(bin_vals) == 0:
                    C_vals.append(np.nan)
                else:
                    C_vals.append(np.nanmean(bin_vals))

            C_vals = np.array(C_vals)

            # --- Skip invalid ---
            if np.any(~np.isfinite(C_vals)) or np.any(C_vals <= 0):
                k_dict[label].append(np.nan)
                continue

            # --- Transform based on reaction order ---
            if reaction_order == 'Zero':
                y = C_vals
            elif reaction_order == 'One':
                y = np.log(C_vals + 1e-12)
            elif reaction_order == 'Two':
                y = 1 / C_vals
            else:
                raise ValueError("reaction_order must be 'Zero', 'One', or 'Two'")

            # --- Fit ---
            slope, intercept = np.polyfit(times, y, 1)
            fit = slope * times + intercept

            # --- R² ---
            ss_res = np.sum((y - fit)**2)
            ss_tot = np.sum((y - np.mean(y))**2)
            r2 = 1 - ss_res / ss_tot if ss_tot > 0 else 0

            # --- Store k correctly ---
            if r2 >= rejection_threshold:

                if reaction_order == 'Zero':
                    k = -slope
                elif reaction_order == 'One':
                    k = slope
                elif reaction_order == 'Two':
                    k = slope   # ⚠️ important: 2nd order slope = +k

                k_dict[label].append(k)

            else:
                k_dict[label].append(np.nan)
       

    

    plt.figure()
    

    bin_widths = np.array([
        0.48, 0.26, 0.31, 0.37, 0.44, 0.52, 0.60, 0.72, 0.85, 1,
        1.17, 1.4, 1.68, 1.9, 2.3, 2.7, 3.2, 3.7, 4.5, 5.2,
        6.2, 7.2, 8.6, 10.2, 11.9, 14.2, 16.4, 20, 23, 28,
        32, 38, 47, 57, 66, 80
    ])

    # Convert widths → edges (log-spaced style assumption)
    edges = np.zeros(len(bin_widths) + 1)
    edges[0] = bin_widths[0] / 2

    for i in range(1, len(edges)):
        edges[i] = edges[i-1] + bin_widths[i-1]

    # Centers
    centers = 0.5 * (edges[:-1] + edges[1:])

    # Asymmetric x-errors
    xerr = np.vstack([
        centers - edges[:-1],   # left
        edges[1:] - centers     # right
    ])

    marker_map = {
        "Baseline": "D",
        "Low": "s",
        "Med": "o",
        "High": "^",
        "Mix": "v"
    }
    # linestyle_map = {
    #     "Baseline": (0, (1, 1)),        # densely dotted
    #     "Low":      (0, (5, 2)),        # dashed
    #     "Med":      "solid",            # solid
    #     "High":     (0, (3, 2, 1, 2)),  # dash-dot
    #     "Mix":      (0, (10, 3))        # long dash
    # }
    color_map = {
        "Baseline": "0.2",   # dark gray
        "Low":      "0.4",
        "Med":      "0.0",   # pure black
        "High":     "0.6",
        "Mix":      "0.8"    # light gray
    }

    for label in group_labels:

        plt.errorbar(
            centers[1:],
            k_dict[label][1:],
            xerr=xerr[:, 1:],

            fmt=marker_map[label],
            linestyle=None,
            color=color_map[label],

            markeredgecolor='black',
            markerfacecolor=color_map[label],
            markeredgewidth=1.2,

            linewidth=1.5,
            capsize=3,

            label=label
        )
        

    plt.xscale('log')
    tick_values = [1, 5, 25, 100, 500]
    plt.xticks(tick_values, labels=[str(v) for v in tick_values])
    plt.xlabel("Particle Bin Center (μm)")

    if reaction_order == 'Zero':
        ylabel = "C"
    elif reaction_order == 'One':
        ylabel = "ln(C)"
    elif reaction_order == 'Two':
        ylabel = "1/C"

    #plt.ylabel(f"k ({reaction_order}-order), based on {ylabel}")
    plt.ylabel("k [1/h]")
    plt.title(title if title else "k vs Bin (by condition)")
    plt.legend()
    plt.grid(True)
    plt.show()

    if bin_id is not None:
        if reaction_order == 'Zero':
            plot_bin_C_vs_time(results_t1, results_t2, results_t3, bin_index=bin_id, title=title)
        elif reaction_order == 'One':
            plot_bin_lnC_vs_time(results_t1, results_t2, results_t3, bin_index=bin_id, title=title)
        elif reaction_order == 'Two':
            plot_bin_1_C_vs_time(results_t1, results_t2, results_t3, bin_index=bin_id, title=title)

    return 

plotk(filepath_t1="lap_t3_rs.csv", filepath_t2="lap_t23_rs.csv", filepath_t3="lap_t27_rs.csv",
      sample_window1=lap_t3, sample_window2=lap_t23, sample_window3=lap_t27,
      title="(b) Laponite", condition=["Low", "Med", "High"], rejection_threshold=0.3, bin_id=5, reaction_order="One")


plotk(filepath_t1="kao_t3_rs.csv", filepath_t2="kao_t22_rs.csv", filepath_t3="kao_t26_rs.csv",
      sample_window1=kao_t3, sample_window2=kao_t22, sample_window3=kao_t26,
      title="(a) Kaolinite", condition=["Low", "Med", "High"], rejection_threshold=0.3, bin_id=0, reaction_order="One")
```

```python
# Figure 7
def plot_two_cdfs(hist1, hist2, label1="Dataset 1", label2="Dataset 2",
                  normalize=False, title="Particle Size Distribution (CDF)"):

    import numpy as np
    import matplotlib.pyplot as plt

    hist1 = np.array(hist1, dtype=float)
    hist2 = np.array(hist2, dtype=float)

    # --- Convert to PDFs if needed ---
    if normalize:
        hist1 = hist1 / np.sum(hist1)
        hist2 = hist2 / np.sum(hist2)

    # --- Compute CDFs ---
    cdf1 = np.cumsum(hist1)
    cdf2 = np.cumsum(hist2)

    if cdf1[-1] > 0:
        cdf1 = cdf1 / cdf1[-1]
    if cdf2[-1] > 0:
        cdf2 = cdf2 / cdf2[-1]

    # --- Bin setup ---
    edges = [1.00, 1.48, 1.74, 2.05, 2.42, 2.86, 3.38, 3.98, 4.70,
             5.55, 6.55, 7.72, 9.12, 10.8, 12.7, 15.0, 17.7,
             20.9, 24.6, 29.1, 34.3, 40.5, 47.7, 56.3, 66.5,
             78.4, 92.6, 109, 129, 152, 180, 212, 250,
             297, 354, 420, 500]

    edges = np.array(edges)
    bin_centers = np.sqrt(edges[:-1] * edges[1:])

    # --- Helper: interpolate D50 ---
    def compute_d50(cdf, x):
        idx = np.searchsorted(cdf, 0.5)
        if idx == 0:
            return x[0]
        if idx >= len(cdf):
            return x[-1]
        x0, x1 = x[idx-1], x[idx]
        y0, y1 = cdf[idx-1], cdf[idx]
        return x0 + (0.5 - y0) * (x1 - x0) / (y1 - y0)

    d50_2 = compute_d50(cdf1, bin_centers)
    d50_1 = compute_d50(cdf2, bin_centers)

    # --- Plot ---
    plt.figure(figsize=(4,4))

    plt.plot(bin_centers, cdf1, linestyle='--', color='black', linewidth=2, label=label1)
    plt.plot(bin_centers, cdf2, linestyle='-',  color='black', linewidth=2, label=label2)

    # --- D50 reference lines ---
    plt.axhline(0.5, color='gray', linestyle='-', linewidth=1)
    plt.axvline(d50_1, color='black', linestyle='-', linewidth=1)
    plt.axvline(d50_2, color='black', linestyle='--',  linewidth=1)

    # --- Small icons at D50 points ---
    plt.plot(d50_1, 0.5, marker='o', color='black', markersize=4)
    plt.plot(d50_2, 0.5, marker='o', color='black', markersize=4)

    # --- Vertical labels (log-scale aware offsets) ---
    # multiply x slightly to shift left/right on log scale
    plt.text(d50_1 * 1.1, 0.5, f"{d50_1:.1f}                             ",
             rotation=90, ha='left', va='top', fontsize=8)
    print(d50_1)
    plt.text(d50_2 * 1, 0.5, f"{d50_2:.1f}                             ",
             rotation=90, ha='right', va='top', fontsize=8)

    plt.xscale('log')

    tick_values = [1, 5, 25, 100, 500]
    plt.xticks(tick_values, labels=[str(v) for v in tick_values])

    plt.xlabel("Particle Size (μm)")
    plt.ylabel("Cumulative Fraction")

    if normalize:
        title += " (Normalized)"

    plt.title(title)
    plt.legend()

    plt.grid(True, which='major', alpha=0.3, linewidth=0.5)
    plt.tight_layout()
    plt.show()


def plot_two_histograms(hist1, hist2, label1="Dataset 1", label2="Dataset 2",
                        normalize=False, title="Particle Size Distribution"):
    import numpy as np
    import matplotlib.pyplot as plt

    hist1 = np.array(hist1, dtype=float)
    hist2 = np.array(hist2, dtype=float)

    # Optional normalization
    if normalize:
        hist1 = hist1 / np.sum(hist1)
        hist2 = hist2 / np.sum(hist2)

    # --- Bin setup (your exact logic preserved) ---
    edges = [1.00, 1.48, 1.74, 2.05, 2.42, 2.86, 3.38, 3.98, 4.70,
             5.55, 6.55, 7.72, 9.12, 10.8, 12.7, 15.0, 17.7,
             20.9, 24.6, 29.1, 34.3, 40.5, 47.7, 56.3, 66.5,
             78.4, 92.6, 109, 129, 152, 180, 212, 250,
             297, 354, 420, 500]

    
    # Create 36 visually uniform log bins from 1 to 500
    plot_edges = np.logspace(np.log10(1), np.log10(500), len(hist1) + 1)

    left_edges = plot_edges[:-1]
    new_widths = np.diff(plot_edges)



    # --- Plot ---
    plt.figure(figsize=(12,4))

    # First histogram
    plt.bar(left_edges, hist1,
            width=new_widths,
            align='edge',
            alpha=0.9,
            label=label1,
            color='white',
            edgecolor='black',
            hatch='///')

    # Second histogram
    plt.bar(left_edges, hist2,
            width=new_widths,
            align='edge',
            alpha=0.5,
            label=label2,
            color='black',
            edgecolor='white'
            )

    plt.xscale('log')

    tick_values = [1, 5, 25, 100, 500]
    plt.xticks(tick_values, labels=[str(v) for v in tick_values])

    plt.xlabel("Particle Size (μm)")
    plt.ylabel("Mean Volume Concentration (μL/L)")

    if normalize:
        title += " (Normalized)"

    plt.title(title)

    plt.legend()

    plt.grid(True, which='major', alpha=0.3, linewidth=0.5)
    plt.tight_layout()
    plt.show()


plot_two_histograms(lt3_slice, lt23_slice,
                    label1="Time = 3h",
                    label2="Time = 23h",
                    normalize=False,
                    title="(c) Laponite Treatment")


plot_two_histograms(kt3_slice, kt22_slice,
                    label1="Time = 3h",
                    label2="Time = 22h",
                    normalize=False,
                    title="(a) Kaolinite Treatment")



plot_two_cdfs(hist1=kt3_slice,
              hist2=kt22_slice,
              label1="Time = 3h",
              label2="Time = 22h",
              normalize=False,
              title="(b) Kaolinite Empirical CDF")

plot_two_cdfs(hist1=lt3_slice,
              hist2=lt23_slice,
              label1="Time = 3h",
              label2="Time = 23h",
              normalize=False,
              title="(d) Laponite Empirical CDF")
```

```python
# Supplementary
lap_voltages = [0.232,
0.223,
0.242,
0.221,
0.23,
0.225,
0.181,
0.183,
0.19,
0.184,
0.192,
0.197,
0.198,
0.206,
0.187,
0.188,
0.194,
0.172,
0.075,
0.07,
0.105,
0.187,
0.184,
0.192,
0.177,
0.198,
0.184,
0.043,
0.057,
0.062,
0.026,
0.028,
0.029,
0.135,
0.13,
0.13,
0.126,
0.288,
0.255,
0.23,
0.162,
0.147,
0.155,
0.154,
0.141,
0.152,
0.143,
0.123]
lap_conc = [5.586,
6.044,
5.642,
5.74,
6.196,
5.372,
5.03,
5.572,
5.378,
5.012,
5.352,
4.934,
6.018,
6.216,
5.676,
5.17,
5.216,
4.496,
1.542,
1.728,
2.778,
6.434,
6.57,
5.714,
5.558,
6.406,
5.058,
0.826,
0.824,
1.31,
0.356,
0.556,
0.246,
5.07,
4.84,
4.72,
4.28,
8.61,
5.88,
6.73,
4.51,
4.43,
4.69,
4.56,
4.72,
5.07,
4.78,
4.38]
kao_voltages = [0.596,
0.573,
0.47,
0.482,
0.466,
0.431,
0.419,
0.309,
0.348,
0.336,
0.238,
0.218,
0.217,
0.183,
0.173]
kao_conc = [2.88,
1.91,
2.09,
2.41,
1.5,
2.07,
2.1,
1.9,
1.67,
1.56,
1.21,
1.36,
1.29,
1.13,
1.2]

def cal_curve(lvolt, lconc, kvolt, kconc):
    """
    Plot calibration curves for Laponite and Kaolinite.

    Parameters
    ----------
    lvolt : array-like
        Laponite LISST voltages

    lconc : array-like
        Laponite chlorophyll concentrations (ug/L)

    kvolt : array-like
        Kaolinite LISST voltages

    kconc : array-like
        Kaolinite chlorophyll concentrations (ug/L)

    Notes
    -----
    - Black-and-white publication style
    - Displays regression equation and R²
    """

    # --- Convert to arrays ---
    lvolt = np.asarray(lvolt)
    lconc = np.asarray(lconc)

    kvolt = np.asarray(kvolt)
    kconc = np.asarray(kconc)

    # --- Linear regressions ---
    lslope, lintercept, lr, _, _ = linregress(lvolt, lconc)
    kslope, kintercept, kr, _, _ = linregress(kvolt, kconc)

    lr2 = lr**2
    kr2 = kr**2

    # --- Fit lines ---
    lx = np.linspace(np.min(lvolt), np.max(lvolt), 200)
    ly = lslope * lx + lintercept

    kx = np.linspace(np.min(kvolt), np.max(kvolt), 200)
    ky = kslope * kx + kintercept

    # --- Plot ---
    plt.figure(figsize=(10,10))

    # Laponite
    plt.scatter(
        lvolt,
        lconc,
        marker='o',
        facecolors='white',
        edgecolors='black',
        s=50,
        label='Laponite data'
    )

    plt.plot(
        lx,
        ly,
        linestyle='--',
        color='black',
        linewidth=1.5,
        label=(
            f"Laponite fit: "
            f"y={lslope:.2f}x+{lintercept:.2f}, "
            f"$R^2$={lr2:.3f}"
        )
    )

    # Kaolinite
    plt.scatter(
        kvolt,
        kconc,
        marker='s',
        facecolors='0.6',
        edgecolors='black',
        s=50,
        label='Kaolinite data'
    )

    plt.plot(
        kx,
        ky,
        linestyle=':',
        color='0.3',
        linewidth=1.5,
        label=(
            f"Kaolinite fit: "
            f"y={kslope:.2f}x+{kintercept:.2f}, "
            f"$R^2$={kr2:.3f}"
        )
    )

    # --- Labels ---
    plt.xlabel("LISST Signal (V)")
    plt.ylabel("Chlorophyll Concentration (μg/L)")

    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # --- Print results ---
    print("Laponite:")
    print(f"  slope = {lslope:.4f}")
    print(f"  intercept = {lintercept:.4f}")
    print(f"  R² = {lr2:.4f}")

    print("\nKaolinite:")
    print(f"  slope = {kslope:.4f}")
    print(f"  intercept = {kintercept:.4f}")
    print(f"  R² = {kr2:.4f}")

    return {
        "laponite": {
            "slope": lslope,
            "intercept": lintercept,
            "r2": lr2
        },
        "kaolinite": {
            "slope": kslope,
            "intercept": kintercept,
            "r2": kr2
        }
    }

cal_curve(
    lap_voltages, lap_conc,
    kao_voltages, kao_conc
)
```
